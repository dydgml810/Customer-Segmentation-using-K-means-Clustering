```python
#   ( )
def detect_outliers(df, columns):
  q_hi = df[columns].quantile(0.75)
  q_low = df[columns].quantile(0.25)
  iqr = q_hi-q_low

  condition_1 = df[columns] > q_hi+1.5*iqr
  condition_2 = df[columns] < q_low-1.5*iqr

  a1=df[condition_1].index
  a2=df[condition_2].index

  df.drop(a1, axis=0, inplace=True)
  df.drop(a2, axis=0, inplace=True)

  return df
```

# K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
#  
def elbow(data, k):
  sse = []
  for i in range(1, k):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)

    sse.append(kmeans.inertia_)
  plt.plot(range(1, k), sse, 'bo-')
  plt.title("elbow method")
  plt.xlabel("number of clusters")
  plt.ylabel("SSE")
  plt.show()
```

```python
#  
def silhouette(data, k):
  sil_score = []
  for i in range(2, k):
    kmeans = KMeans(n_clusters=i)
    cluster_labels = kmeans.fit_predict(data)

    silhouette_avg = silhouette_score(data, cluster_labels)
    sil_score.append(silhouette_avg)

  plt.plot(range(2, k), sil_score, 'bo-')
  plt.title("silhouette")
  plt.xlabel("number of clusters")
  plt.ylabel("silhouette_coef")
  plt.show()
```

```python
elbow(df_pca_0, 21)
```

```python
silhouette(df_pca_0, 21)
```

```python
elbow(df_pca_1, 15)
```

```python
silhouette(df_pca_1, 20)
```
