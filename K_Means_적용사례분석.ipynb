```python
# 이상치 제거 (사분위수 기준)
def detect_outliers(df, columns):
  q_hi = df[columns].quantile(0.75)
  q_low = df[columns].quantile(0.25)
  iqr = q_hi-q_low

  condition_1 = df[columns] > q_hi+1.5*iqr
  condition_2 = df[columns] < q_low-1.5*iqr

  a1=df[condition_1].index
  a2=df[condition_2].index

  df.drop(a1, axis=0, inplace=True)
  df.drop(a2, axis=0, inplace=True)

  return df
```

# K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
# 엘보우 기법
def elbow(data, k):
  sse = []
  for i in range(1, k):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)

    sse.append(kmeans.inertia_)
  plt.plot(range(1, k), sse, 'bo-')
  plt.title("elbow method")
  plt.xlabel("number of clusters")
  plt.ylabel("SSE")
  plt.show()
```

```python
# 실루엣 기법
def silhouette(data, k):
  sil_score = []
  for i in range(2, k):
    kmeans = KMeans(n_clusters=i)
    cluster_labels = kmeans.fit_predict(data)

    silhouette_avg = silhouette_score(data, cluster_labels)
    sil_score.append(silhouette_avg)

  plt.plot(range(2, k), sil_score, 'bo-')
  plt.title("silhouette")
  plt.xlabel("number of clusters")
  plt.ylabel("silhouette_coef")
  plt.show()
```

```python
elbow(df_pca_0, 21)
```

```python
silhouette(df_pca_0, 21)
```

```python
elbow(df_pca_1, 15)
```

```python
silhouette(df_pca_1, 20)
```

```python
# K-Means 군집화 실행(st_0)

kmeans=KMeans(n_clusters=14, max_iter=200, random_state=0)
cluster_labels = kmeans.fit_predict(df_pca_0)
df_pca_0["kmeans_label"] = cluster_labels
st_0['kmeans_label']=cluster_labels

unique_labels = np.unique(cluster_labels)

for cluster in unique_labels:
  # 각 군집 시각화
  cluster_v = df_pca_0[df_pca_0['kmeans_label'] == cluster]
  plt.scatter(x=cluster_v['component_0'], y=cluster_v['component_1'], edgecolor='k')
plt.show()
```

```python
# K-Means 군집화 실행(st_1)

kmeans=KMeans(n_clusters=8, max_iter=200, random_state=0)
cluster_labels = kmeans.fit_predict(df_pca_1)
df_pca_1["kmeans_label"] = cluster_labels
st_1['kmeans_label']=cluster_labels

unique_labels = np.unique(cluster_labels)

for cluster in unique_labels:
  # 각 군집 시각화
  cluster_v = df_pca_1[df_pca_1['kmeans_label'] == cluster]
  plt.scatter(x=cluster_v['component_2'], y=cluster_v['component_3'], edgecolor='k')
plt.show()
```
